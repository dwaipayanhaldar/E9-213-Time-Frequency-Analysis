```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, fftshift
from scipy.ndimage import convolve
```

```python
# Defining the Box function
def rect(N=200):
    n = np.arange(-N//2, N//2)
    x = np.where(np.abs(n) <= 10, 1, 0)
    return n, x
```

```python
n, x_box = rect()
y_sinc_fft = fft(x_box)
y_sinc_fft = fftshift(y_sinc_fft)
t_fft = fftshift(fftfreq(len(n), d=(n[1]-n[0])))

x_tri = np.convolve(x_box, x_box, mode = 'same')/len(x_tri)
y_sinc2_fft = fft(x_tri)
y_sinc2_fft = fftshift(y_sinc2_fft)

plt.figure(figsize=(16,12))

plt.subplot(2,2,1)
plt.plot(n, x_box, color = 'blue')
plt.xlabel('time(t)')
plt.ylabel('f(t)')
plt.title('Time Domain Signal')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5)

plt.subplot(2,2,2)
plt.plot(t_fft, np.abs(y_sinc_fft)/len(y_sinc_fft), color='red')
plt.xlabel('Frequency(Hz)')
plt.ylabel('|f\u0302(\u03C9 \u2080)|')
plt.title(f'Magnitude Spectrum (FFT)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5) 

plt.subplot(2,2,3)
plt.plot(n, x_tri, color = 'blue')
plt.xlabel('time(t)')
plt.ylabel('f(t)')
plt.title('Time Domain Signal')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5)

plt.subplot(2,2,4)
plt.plot(t_fft, np.abs(y_sinc2_fft)/len(y_sinc2_fft), color='red')
plt.xlabel('Frequency(Hz)')
plt.ylabel('|f\u0302(\u03C9 \u2080)|')
plt.title(f'Magnitude Spectrum (FFT)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5) 


```
